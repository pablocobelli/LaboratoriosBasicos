```python
def GraficarHistogramas(datos):
    figura, sparr = plt.subplots(3, 7, figsize=(15,10), sharex=False, sharey=False)
    sparr = sparr.ravel(order='F')  
    
    for idx, ax in enumerate(sparr):
        if idx != 15:
            ax.hist(datos.Periodo[datos.index==idx].values[0], edgecolor='white', facecolor='C'+str(datos.Grupo[idx]-1))
            ax.set_title(datos.Nombre[datos.index==idx].values[0].upper());
        else:
            ax.set_title(datos.Nombre[datos.index==idx].values[0].upper());
            ax.set_xticks([])
            ax.set_yticks([])
            ax.plot([0, 1], [0, 1], 'black')
            ax.plot([0, 1], [1, 0], 'black')

    plt.tight_layout()
    
GraficarHistogramas(datos)
```

El histograma de 'Lucio' aparece tachado dado que ese grupo (en color marron en la figura) solo midio 2 series de mediciones. 

Tambien resulta interesante graficar los datos 'crudos', para detectar anomalias

```python
def GraficarDatosCrudos(datos):
    figura, sparr = plt.subplots(3, 7, figsize=(15,10), sharex=False, sharey=False)
    sparr = sparr.ravel(order='F')
    
    for idx, ax in enumerate(sparr):
        if idx != 15:
            mediciones = datos.Periodo[datos.index==idx].values[0]
            media = np.mean(mediciones)*np.ones_like(mediciones)
            devst = np.std(mediciones)*np.ones_like(mediciones)
            ax.plot(mediciones, '.-', color='C'+str(datos.Grupo[idx]-1))
            ax.plot(media, '--', color='black')
            ax.plot(media+devst, ':', color='gray')
            ax.plot(media-devst, ':', color='gray')
            ax.set_title(datos.Nombre[datos.index==idx].values[0].upper());
        else:
            ax.set_title(datos.Nombre[datos.index==idx].values[0].upper());
            ax.set_xticks([])
            ax.set_yticks([])
            ax.plot([0, 1], [0, 1], 'black')
            ax.plot([0, 1], [1, 0], 'black')
            
    plt.tight_layout()
    
GraficarDatosCrudos(datos)
```

## Calculemos media, desviacion estandar y distribucion de cada set de datos

Un ejemplo de como calcular y mostrar en pantalla la media y el desvio estandar asociados a las observaciones realizadas por el primer miembro de la Tabla 'datos' (ver tabla mas arriba):

```python
print(np.mean(datos.Periodo[0]))
print(np.std(datos.Periodo[0]))
```

```python
def AnalisisProporcion(datos,n):
    '''
    Calcula la proporcion de datos entre (media-n*sigma) y (media+n*sigma).
    '''
    prop = np.zeros(21)
    for ii in range(0,21):
        mediciones = datos.Periodo[ii]
        nro_meds   = len(mediciones)
    
        media  = np.mean(mediciones)
        desvio = np.std(mediciones)
        cantidad_en_sigma = ( ( mediciones > media-n*desvio ) & ( mediciones < media+n*desvio ) ).sum()
        proporcion_en_sigma = cantidad_en_sigma/nro_meds
    
        prop[ii] = proporcion_en_sigma
    
    return prop
    
```

```python
prop1s = AnalisisProporcion(datos,1)
prop2s = AnalisisProporcion(datos,2)
prop3s = AnalisisProporcion(datos,3)
```
